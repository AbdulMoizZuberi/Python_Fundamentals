#### Exercise 1 - rewrite the above example code using list comprehension syntax. Make a variable named uppercased_fruits to hold the output of the list comprehension. Output should be ['MANGO', 'KIWI', etc...]

### Implement a function to generate random even numbers.

````python
import random

def rand_even_no():
  rand_even_numbers = []

  for _ in range(15):
    num = random.randint(0, 100)
    if num % 2 == 0:
      rand_even_numbers.append(num)

  for i in rand_even_numbers:
    print(i)

rand_even_no()
````

## Exercise 3
### Write a function that uses a loop to calculate the factorial of a number.

````python
def factorial(num):
  if num < 0:
    print("No factorial defined for negative number!")
  else:
    f = 1
    for i in range(1, num + 1):
      f *= i
    print(f)

factorial(8)
````
